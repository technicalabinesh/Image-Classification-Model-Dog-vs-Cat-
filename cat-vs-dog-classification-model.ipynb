```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# ============================
# STEP 1: LOAD FROM TFDS
# ============================
IMG_SIZE = (64, 64)
BATCH    = 32

(raw_train, raw_test), info = tfds.load(
    "cats_vs_dogs",
    split=["train[:80%]", "train[80%:]"],   # 80/20 split (no separate test set in this dataset)
    as_supervised=True,
    with_info=True
)

print("Train samples:", info.splits["train"].num_examples)

# ============================
# STEP 2: PREPROCESS & AUGMENT
# ============================
def preprocess(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

def augment(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
    image = tf.image.random_saturation(image, lower=0.8, upper=1.2)

    # Zoom via crop + resize (equivalent to zoom_range=0.2)
    crop_size = tf.random.uniform((), minval=int(64*0.8), maxval=64, dtype=tf.int32)
    image = tf.image.random_crop(image, size=[crop_size, crop_size, 3])
    image = tf.image.resize(image, IMG_SIZE)   # resize back to 64x64

    image = tf.clip_by_value(image, 0.0, 1.0)  # keep pixels in valid range
    return image, label

train_data = (
    raw_train
    .map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    .map(augment,    num_parallel_calls=tf.data.AUTOTUNE)
    .shuffle(1000)
    .batch(BATCH)
    .prefetch(tf.data.AUTOTUNE)   # loads next batch while GPU trains
)

test_data = (
    raw_test
    .map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(BATCH)
    .prefetch(tf.data.AUTOTUNE)
)

# ============================
# STEP 3: BUILD CNN MODEL
# ============================
model = models.Sequential([
    # Block 1
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(64, 64, 3)),
    layers.BatchNormalization(),          # ← added: stabilizes training
    layers.MaxPooling2D(2, 2),

    # Block 2
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2, 2),

    # Block 3
    layers.Conv2D(128, (3,3), activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2, 2),

    # Classifier head
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
model.summary()

# ============================
# STEP 4: CALLBACKS
# ============================
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=3, restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.5, patience=2, verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        "best_model.keras", save_best_only=True, monitor="val_accuracy"
    )
]

# ============================
# STEP 5: TRAIN
# ============================
history = model.fit(
    train_data,
    epochs=15,
    validation_data=test_data,
    callbacks=callbacks
)

# ============================
# STEP 6: EVALUATE
# ============================
loss, acc = model.evaluate(test_data)
print(f"\nTest Accuracy: {acc:.4f}")

# ============================
# STEP 7: PLOT TRAINING CURVES
# ============================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history["accuracy"],     label="Train Acc")
ax1.plot(history.history["val_accuracy"], label="Val Acc")
ax1.set_title("Accuracy"); ax1.legend()

ax2.plot(history.history["loss"],     label="Train Loss")
ax2.plot(history.history["val_loss"], label="Val Loss")
ax2.set_title("Loss"); ax2.legend()

plt.tight_layout()
plt.savefig("training_curves.png")
plt.show()

# ============================
# STEP 8: PREDICT FROM UPLOADED IMAGE
# ============================
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import ipywidgets as widgets
from IPython.display import display, clear_output
import io

def predict_from_image(img):
    """Accepts a PIL Image object and returns prediction."""
    img_resized   = img.resize(IMG_SIZE)
    img_array     = np.array(img_resized) / 255.0
    img_array     = np.expand_dims(img_array, axis=0)

    prob  = model.predict(img_array, verbose=0)[0][0]
    label = "Dog 🐶" if prob > 0.5 else "Cat 🐱"
    conf  = prob if prob > 0.5 else 1 - prob

    # Show image + result
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"{label}  |  Confidence: {conf:.2%}", fontsize=14, fontweight="bold")
    plt.tight_layout()
    plt.show()

    print(f"Prediction : {label}")
    print(f"Confidence : {conf:.2%}")

# ============================
# UPLOAD WIDGET (Kaggle/Colab)
# ============================
upload = widgets.FileUpload(accept="image/*", multiple=False)
btn    = widgets.Button(description="Predict", button_style="success")
out    = widgets.Output()

def on_predict(change):
    with out:
        clear_output()
        # Works for both old and new ipywidgets versions
        uploaded_files = upload.value
        if not uploaded_files:
            print("Please upload an image first.")
            return

        # ipywidgets >= 8: value is a dict keyed by filename
        if isinstance(uploaded_files, dict):
            file_info = next(iter(uploaded_files.values()))
            content   = file_info["content"]
        else:
            # ipywidgets < 8: value is a list/tuple
            content = uploaded_files[0]["content"]

        img = Image.open(io.BytesIO(bytes(content))).convert("RGB")
        predict_from_image(img)

btn.on_click(on_predict)

print("📁 Upload a cat or dog image and click Predict:")
display(widgets.VBox([upload, btn, out]))
```
